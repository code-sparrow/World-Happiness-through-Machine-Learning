```python
# Importing the libraries
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
df = pd.read_csv('resources/train.csv')
```

```python
X = df.iloc[:, 3:].values
y = df.iloc[:, 2].values
```

```python
index = [(0, 0, 0, "GDP"), (0, 1, 1, "Support"), (1, 0, 2, "Health"), \
         (1, 1, 3, "Freedom"), (2, 0, 4, "Generosity"), (2, 1, 5, "Corruption")]

fig, axs = plt.subplots(3, 2, figsize=(16,16))

for i, j, k, n in index:
    axs[i, j].scatter(X[:,k], y)
    axs[i, j].set_title(n)
    axs[i, j].set(xlabel=n, ylabel="Ladder")
plt.tight_layout()
#plt.show()   
plt.savefig("resources/Simple_Relationships.png", dpi=500)
```
