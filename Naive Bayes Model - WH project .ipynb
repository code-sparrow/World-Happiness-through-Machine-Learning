```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import os
```

```python
happydata = pd.read_csv("../Resources/WHR20_Data.csv", usecols=['Country', 'year', 'Ladder', 'GDP_per_capita', 'Social_support', 'Healthy_life_expectancy_at_birth', 'Freedom_to_make_life_choices', 'Generosity','Perceptions_of_corruption'])
happydata.head(25)                  
```

```python
print(happydata.shape)
```

```python
happy_df = happydata.dropna()
happy_df
```

```python
happy_df.describe()
```

```python
#scroll right, basically changing the data to 0s and 1s, the one hot encoding
happy_df = pd.get_dummies(happy_df)
happy_df.iloc[:,:].head(15)
```

```python
labels = np.array(happy_df["Ladder"])
happy_df = happy_df.drop("Ladder", axis= 1)
happy_df_list = list(happy_df.columns)
happy_df = np.array(happy_df)
```

```python
from sklearn.model_selection import train_test_split
train_happy_df, test_happy_df, train_labels, test_labels = train_test_split(happy_df, labels, test_size = 0.25, random_state = 42)
```

```python
print(train_happy_df.shape)
print(train_labels.shape)
print(test_happy_df.shape)
print(test_labels.shape)
```

```python
# Support vector machine linear classifier
from sklearn import linear_model
model = linear_model.BayesianRidge()

model.fit(train_happy_df, train_labels)

```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(test_happy_df, test_labels))
```
