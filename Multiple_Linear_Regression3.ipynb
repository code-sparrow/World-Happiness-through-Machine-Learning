```python
# Multiple Linear Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
df_train = pd.read_csv('resources/train.csv')
df_test = pd.read_csv('resources/test.csv')
```

```python
df_train.head()
```

```python
print("Train: ", df_train.shape)
print("Test: ", df_test.shape)
```

```python
# Setting up arrays of predictors and target
X_train = df_train.iloc[:, 3:].values
y_train = df_train.iloc[:, 2].values
X_test = df_test.iloc[:, 3:].values
y_test = df_test.iloc[:, 2].values
```

```python
# Fitting Multiple Linear Regression to the Training set (all predictors)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Using all predictors R2
regressor.score(X_test, y_test)
```

```python
# Building the optimal model using Backward Elimination
import statsmodels.api as sm
# Need to add a column of ones at the beginning of X-array for the constant in the linear model
X_train2 = np.append(arr = np.ones((len(y_train), 1)).astype(int), values = X_train, axis = 1)
# Selecting all features/predictors (and new one's column)
X_opt = X_train2[:, [0, 1, 2, 3, 4, 5, 6]]
regressor_OLS = sm.OLS(endog = y_train, exog = X_opt).fit()
regressor_OLS.summary()
```
